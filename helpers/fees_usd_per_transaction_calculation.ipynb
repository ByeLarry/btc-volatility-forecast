```python
import pandas as pd
import numpy as np

transactions = pd.read_csv(
    "data/csv/n-transactions.csv",
    header=None,
    names=["date", "n_transactions"]
)

fees = pd.read_csv(
    "data/csv/transaction-fees-usd.csv",
    header=None,
    names=["date", "transaction_fees_usd"]
)

transactions["date"] = pd.to_datetime(transactions["date"])
fees["date"] = pd.to_datetime(fees["date"])

df = transactions.merge(fees, on="date", how="inner")

df["fees_usd_per_transaction"] = np.where(
    df["n_transactions"] > 0,
    df["transaction_fees_usd"] / df["n_transactions"],
    0
)

result = df[["date", "fees_usd_per_transaction"]]

result.to_csv(
    "fees-usd-per-transaction.csv",
    index=False,
    header=False
)

print("Файл fees-usd-per-transaction.csv создан")
print(result.head())
```
